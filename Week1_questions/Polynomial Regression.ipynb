## 1. Polynomial Regression 
### 1) 다항회귀란?   
- 비선형 데이터를 학습하기 위해 선형모델을 사용하는 기법으로, 각 변수의 거듭제곱을 새로운 변수로 추가하고, 이 확장된 변수를 포함한 데이터셋을 가지고 선형모델을 훈련시키는 방법이다. 

### 2) 다항회귀에서 다항인 부분?
- 독립변수가 단항식이 아닌 (feature1, feature2, ...) 독립변수가 2차, 3차로 표현되는 것을 다항회귀라고 한다. (feature1, feature1^2, ...) 
- 회귀분석에서 선형/비선형 회귀의 구분은 독립변수가 아닌 회귀계수(w_0, w_1...)의 선형성 여부를 의미하는 것으로, 회귀 계수는 선형 이면서 독립변수가 다항이므로 다항선형 회귀라고 한다. 
- 사이킷런은 직접적으로 다항회귀를 구현해주지 않기 떄문에 PolynomialFeatures라는 클래스를 통해 피처를 다항식으로 우선 변경 후, 선형 회귀를 적용해 주어야 한다.

## 2. Python으로 구현해보기

### 1) 2차 방정식 활용 비선형 데이터 생성 

```python
import numpy as np
import matplotlib.pyplot as plt

m = 100
X = 6 * np.random.rand(m,1) - 3
y = 0.5 * X**2 + X + 2 + np.random.randn(m,1)  # 약간의 노이즈 포함

plt.plot(X,y,"b.")
plt.ylabel("Y", fontsize=15,rotation=0)
plt.xlabel("X", fontsize=15)
plt.show()
```

- 그림과 같은 비선형 데이터의 경우 선형모델을 통해 예측하게 되면 제대로 된 예측 모델을 구축할 수가 없다. 따라서 다항회귀를 통해 예측이 필요하다. 

### 2)  사이킷런의 PolynomialFeatures
- 사이킷런의 PolynomialFeatures 가 내장되어 있으며 이를 통해 input 데이터를 활용하여 다항식을 생성할 수 있다.

-  PolynomialFeatures(degree=2, include_bias=False) 설명   
   a. degree=2 : 2차다항을 생성   
   b. include_bias=True : 편향 추가  
   c. interection_only=True : 거듭제곱인 항은 제외 (교차항만 포함) 
- get_feature_names() : 생성된 변수의 차수 확인 가능 

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

print(X[0])
print(X_poly[0])
```

- 변수(X)값들을 제곱하여(degree=2), 새로운 변수를 만들어 주는 역할을 한다.
- 결과 :   
  a. X[0] : X[0]에 해당하는 값   
  b. X_poly[0] : X[0], X[0]^2값   

### 3) Regression 

- Polynomial을 활용, 훈련데이터 변수가 1개에서 2개가 되었고, 이 데이터를 활용하여 LinearRegression을 적용해보면 

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X_poly, y)
lin_reg.intercept_ , lin_reg.coef_
```

```python
X_new=np.linspace(-3, 3, 100).reshape(100, 1)
X_new_poly = poly_features.transform(X_new)
y_new = lin_reg.predict(X_new_poly)
plt.plot(X, y, "b.")
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.legend(loc="upper left", fontsize=14)
plt.axis([-3, 3, 0, 10])
plt.show()
```

- PolynomialFeatures을 사용하면 위와 같이 거듭제곱 항들이 생성되며 독립변수들 간의 교차항도 생성 시킬 수 있다. 
따라서 PolynomialFeatures(degree=d)는 
변수가 n개인 배열의 변수일 때 (n+d)! / d!n!개의 항들이 추가되므로 너무 늘어나지 않게 주의해야한다. 

- degree(차수)가 너무 높아질 경우, 과대적합 발생 

```python
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

for style, width, degree in (("g-", 1, 300), ("b--", 2, 2), ("r-+", 2, 1)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = Pipeline([
            ("poly_features", polybig_features),
            ("std_scaler", std_scaler),
            ("lin_reg", lin_reg),
        ])
    polynomial_regression.fit(X, y)
    y_newbig = polynomial_regression.predict(X_new)
    plt.plot(X_new, y_newbig, style, label=str(degree), linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper left")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([-3, 3, 0, 10])
plt.show()
```

- 위와 같이 과대적합이 발생하므로 cross_val_score등을 통해 검증 필요

```python
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

def plot_learning_curves(model, X, y):
    X_train, X_val, y_train, y_val = train_test_split(X,y, test_size=0.2)
    train_errors, val_errors = [], []
    for m in range (1, len(X_train)):
        model.fit(X_train[:m], y_train[:m])
        y_train_predict = model.predict(X_train[:m])
        y_val_predict = model.predict(X_val)
        train_errors.append(mean_squared_error(y_train[:m], y_train_predict))
        val_errors.append(mean_squared_error(y_val, y_val_predict))
    plt.plot(np.sqrt(train_errors), "r-+", linewidth = 2, label = "train set")
    plt.plot(np.sqrt(val_errors), "b-", linewidth = 3, label = "validation set") 
    plt.xlabel("size of train set")
    plt.ylabel("RMSE")
    plt.legend()
    
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

polynomial_regression = Pipeline([
    ("poly_features", PolynomialFeatures(degree=20, include_bias = False)),
    ("lin_reg", LinearRegression())
])

plot_learning_curves(polynomial_regression, X, y)
plt.axis([0,80,0,3])
plt.show()
```

#### 1. 편향
- 편향은 잘못된 가정으로 인한 것으로, 편향이 큰 모델은 훈련데이터에 과소적합되기 쉽다.(예를들어 비선형에 선형회귀 적합)
- 즉, 직선으로 데이터를 표현하는 경우, 데이터를 충분히 표현해주지 못하게 되며 이로 인해 발생하는 차이가 편향을 뜻한다. 

#### 2. 분산(variance)
- 훈련데이터에 있는 작은 변동에 모델이 과도하게 민감하기 때문에 발생, 자유도가 높은 모델(예를들어 고차 다항회귀)이 높은 분산을 가지기 쉬워 훈련데이터에 과대적합 되는 경향이 있다.
- 즉, 다항 회귀가 되면 차수가 너무 크게 되어, 변동성이 커지게 되며 이를 고분산성을 가진다고 이야기 한다. 

#### 3. 줄일 수 없는 오차(irreducible error)
- 데이터 자체에 있는 노이즈로 인한 오차, 노이즈를 제거하는 방법밖에 없다.


<B><I> 따라서 편향과 분산은 어느 한쪽이 좋아지면 어느 한쪽은 낮아지는 트레이트-오프관계를 가지며 이 둘의 성능을 적절하게 맞춰 오류가 가장 낮아지는 지점을 찾는 것이 가장 중요하다.</B></I>
